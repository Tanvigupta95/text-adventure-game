```python
'''This code is created by Tanvi Gupta
This game is based on my life itself as i am a mountaineer and have climbed
two mountains in my life in the Himalayan region of India. Come and experience 
the greatest adventure and aderaline rush you feel while climbing the highest 
mountain on earth. I hope you enjoy it.
Submitted To - Prof. Chase Kusterer


'''
import time #importing time function
import sys  #importing system function
a=['A','a'] #assigning a for user input
b=['B','b'] #assigning b for user input
yes=['y','Y','yes','Yes','YES'] #assigning yes as global function
no=['n','N','no','No','NO']#assigning no as global function
name='0'
chances=2
def try_again(): #defining try again function
    print('Please enter a valid option:')
def intro():#defining intro
    print('''In this game we will be climbing the highest mountain on planet earth 
    located in the Himalayan mountain range in Nepal together to reach to the 
    summit point at 29029 feet.  In this game you will be climbing the snowy 
    cliffs surrounded by frozen river, magnificent landscapes. You will be encountering
    uncertain situations while you ascend. This game needs amount of courage and
    bravery to make your way to the highest point. You will be embracing your 
    inner beast, you take pride in your survival skills''')
    play=input(prompt='Would you like to play (yes/no)')#user input 
    if play in yes:#using if function
        print('Lets begin.....')
    elif play in no:
        print('Alright big guy suit yourself')
        sys.exit()#calling system exit function
    else:
        try_again()#calling try again function
        intro()
intro()
#Level 1
name=input(prompt='Enter your name please:')#user input
def level_1(): #defining level 1
    global name #calling global function name
    global chances #calling global function chances
    print(f'''{name}, You reached everest base camp at 17600 ft with five other 
          team members who are doing this expedition with you \U0001F3D4. On
          your way to climb the advance base camp you will need to use a ladder
          to cross a huge crevasse in the Khumbu icefall. While crossing the ladder, 
          you suddenly lost control because of the wind. What will you do next?
          A)	Will you  continue to make balance and keep crossing
            B)	Will you get scared and hold the ladder ''')
    choice_1=input(prompt='Choose an option:')#user input
    if choice_1 in a:
        if chances>0:
            print('Yes! You continue your journey.....')
        else:
            print('You are out of lives')
            sys.exit()
    elif choice_1 in b:
        chances-=1
        if chances > 0:
            print(f'You have {chances} chances left. Retry...')
            level_1()
        else:
            print('You are out of lives')
            sys.exit()
    else:
        try_again()
        level_1()
level_1()
time.sleep(2)#time delay
def level_2(): #defining level 2
    global chances
    print('''After crossing the ladder successfully, you reached the advance base camp.
        Now its time for you to up your game and start ascending towards the camp 2 .
        With the next dawn you start climbing again and witness avalanche warning signs.  
        You hear a “whumping” sound as you climb which shows signs that snow is settling 
        and a slab might release. How will you overcome this situation?
        A)	Outrun the avalanche 
        B)	Breathe calmly and try to start climbing towards the safer direction''')
    choice_2=input(prompt='Choose an option:')
    if choice_2 in a:
        chances-=1
        if chances > 0:
            print(f'You have {chances} chances left. Retry...')
            level_2()
        else:
            print('You are out of lives')
            sys.exit()
    elif choice_2 in b:
        if chances>0:
            print('Yes! You continue your journey.....')
        else:
            print('You are out of lives')
            sys.exit()
    else:
        try_again()
        level_2()
level_2()
time.sleep(2)
def level_3(): #defining level 3
    global chances
    print('''Nice choice. After making your way to camp 2 and overcoming the 
    greatest fear of any mountaineer of coming across avalanche and coming out 
    alive, you should be proud of yourself. Your next goal is to reach camp 3 
    successfully at 25,600 ft and acclimatize yourself to be fit for summit. 
    While climbing you suffer from high altitude hallucinations about  a snow 
    leapord \U0001F406 coming towards you. What would you do?
    A)	Attack the snow leapord
    B)	Will you ask base camp through satellite phone to connect you with your family ''')
    choice_3=input(prompt='Choose an option:') 
    if choice_3 in a:
        chances-=1
        if chances > 0:
            print(f'You have {chances} chances left. Retry...')
            level_3()
        else:
            print('You are out of lives')
            sys.exit()
    elif choice_3 in b:
        if chances>0:
            print('Yes! You continue your journey.....')
        else:
            print('You are out of lives')
            sys.exit()
    else:
        try_again()
        level_3()
level_3()
time.sleep(2)
def level_4(): #defining level 4
    global chances
    print('''You made it so far. Now just few more steps remaining between you
    and the summit. After spending two nights at camp 2, one night at camp 3, 
    you will be spending a short night at south cole before climbing for the 
    day you have been waiting for. You have limited  oxygen to support the 
    climbing for the summit.  What would you do?
        a)	Would you still go ahead with the less oxygen 
        b)	Would you wait for someone from your team to bring you oxygen from camp 3''')
    choice_4=input(prompt='Choose an option:')
    if choice_4 in a:
        chances-=1
        if chances > 0:
            print(f'You have {chances} chances left. Retry...')
            level_4()
        else:
            print('You are out of lives')
            sys.exit()
        sys.exit()
    elif choice_4 in b:
        if chances>0:
            print('Yes! You win!')
        else:
            print('You are out of lives')
            sys.exit()
    else:
        try_again()
        level_4()
level_4()
    
    
```
